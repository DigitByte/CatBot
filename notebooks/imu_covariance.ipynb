```python
import rospy
from sensor_msgs.msg import Imu
import numpy as np
import matplotlib.pyplot as plt
from tqdm import tqdm
```

CatBot Lab Context: This notebook supports the CatBot course and focuses on Measure IMU sensor.

# Measure IMU sensor

rosrun rosserial_python serial_node.py /dev/ttyACM0

```python



max_samples = 1000

rospy.init_node('imuData_analysis', anonymous=True)
lin_acc  = np.zeros([3, max_samples])
ang_vel  = np.zeros([3, max_samples])
quat_rot = np.zeros([4, max_samples])

for i in tqdm(range(max_samples)):
    
    msg = rospy.wait_for_message('/catbot/imu_data', Imu, 10)
    
    quat_rot[0,i] = msg.orientation.x
    quat_rot[1,i] = msg.orientation.y
    quat_rot[2,i] = msg.orientation.z
    quat_rot[3,i] = msg.orientation.w
    
    ang_vel[0,i] = msg.angular_velocity.x
    ang_vel[1,i] = msg.angular_velocity.y
    ang_vel[2,i] = msg.angular_velocity.z
    
    lin_acc[0,i] = msg.linear_acceleration.x
    lin_acc[1,i] = msg.linear_acceleration.y
    lin_acc[2,i] = msg.linear_acceleration.z

```

```python
np.savetxt('data/imu_data_orientation.txt',         quat_rot)
np.savetxt('data/imu_data_angular_velocity.txt',    ang_vel)
np.savetxt('data/imu_data_linear_acceleration.txt', lin_acc)
```

```python
quat_rot = np.loadtxt('data/imu_data_orientation.txt')
ang_vel  = np.loadtxt('data/imu_data_angular_velocity.txt')
lin_acc  = np.loadtxt('data/imu_data_linear_acceleration.txt')
```

```python
quat_cov_matrix = np.cov(quat_rot)
print(quat_cov_matrix)
```

# Linear acceleration covariance matrix

```python
lin_acc_matrix = np.cov(lin_acc)
print(lin_acc_matrix)
np.savetxt('data/lin_acc_cov.txt', lin_acc_matrix)
```
