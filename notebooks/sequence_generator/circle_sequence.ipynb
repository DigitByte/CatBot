```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def getPosition(ix_pos, deltat, base_pos, base_angle, feet_pos, stance = [1,1,1,1]):
    
    txt = '[position {:d}] \n'.format(ix_pos)
    txt+= 'time_to_complete   = {:.3f}\n'.format(deltat)
    txt+= 'base_pos_x         = {:.3f}\n'.format(base_pos[0])
    txt+= 'base_pos_y         = {:.3f}\n'.format(base_pos[1])
    txt+= 'base_pos_z         = {:.3f}\n'.format(base_pos[2])
    txt+= 'base_angle_roll  = {:.3f}\n'.format(base_angle[0])
    txt+= 'base_angle_pitch = {:.3f}\n'.format(base_angle[1])
    txt+= 'base_angle_yaw   = {:.3f}\n'.format(base_angle[2])
    txt+= 'stance_fl = {:d}\n'.format(stance[0])
    txt+= 'stance_fr = {:d}\n'.format(stance[1])
    txt+= 'stance_bl = {:d}\n'.format(stance[2])
    txt+= 'stance_br = {:d}\n'.format(stance[3])
    txt+= 'feet_pos_fl_x = {:.3f}\n'.format(feet_pos[0])
    txt+= 'feet_pos_fl_y = {:.3f}\n'.format(feet_pos[1])
    txt+= 'feet_pos_fl_z = {:.3f}\n'.format(feet_pos[2])
    txt+= 'feet_pos_fr_x = {:.3f}\n'.format(feet_pos[0])
    txt+= 'feet_pos_fr_y = {:.3f}\n'.format(-feet_pos[1])
    txt+= 'feet_pos_fr_z = {:.3f}\n'.format(feet_pos[2])
    txt+= 'feet_pos_bl_x = {:.3f}\n'.format(-feet_pos[0])
    txt+= 'feet_pos_bl_y = {:.3f}\n'.format(feet_pos[1])
    txt+= 'feet_pos_bl_z = {:.3f}\n'.format(feet_pos[2])
    txt+= 'feet_pos_br_x = {:.3f}\n'.format(-feet_pos[0])
    txt+= 'feet_pos_br_y = {:.3f}\n'.format(-feet_pos[1])
    txt+= 'feet_pos_br_z = {:.3f}\n'.format(feet_pos[2])

    
    return txt


```

CatBot Lab Context: This notebook supports the CatBot course and focuses on Horizontal circle.

# Horizontal circle

```python
RADIUS_CIRCLE = 0.05 #m
TOTAL_TIME    = 5 #sec
N_STEPS       = 100

prev_pos =  [0.05,0,0]
print(getPosition(0, 0.5, prev_pos, [0,0,0], [0,0.03,-0.15]))


time_arr = np.linspace(0, TOTAL_TIME, N_STEPS)
x = RADIUS_CIRCLE*np.cos(time_arr*2*np.pi/TOTAL_TIME)
y = RADIUS_CIRCLE*np.sin(time_arr*2*np.pi/TOTAL_TIME)

for t in range(1,len(time_arr)):
    dt = time_arr[t] - time_arr[t-1]
    
    current_pos = np.array([x[t], y[t], 0])
    dpos = current_pos - prev_pos
    
    print(getPosition(t, dt, dpos, [0,0,0], [0,0.03,-0.15]))
    
    prev_pos = current_pos

    
plt.plot(x, y)

final_pos = np.array([0,0,0])
dpos = final_pos - prev_pos

print(getPosition(t+1, 0.5, dpos, [0,0,0], [0,0.03,-0.15]))
```
